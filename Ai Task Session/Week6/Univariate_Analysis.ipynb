```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Right Skewed Data (Positive Skew)
right_skewed = [2, 3, 4, 5, 6, 20]

# Left Skewed Data (Negative Skew)
left_skewed = [1, 10, 11, 12, 13]

# Symmetric Data
symmetric = [5, 6, 7, 8, 9]

# Create DataFrame
data = pd.DataFrame({
    "Right Skewed": pd.Series(right_skewed),
    "Left Skewed": pd.Series(left_skewed),
    "Symmetric": pd.Series(symmetric)
})

# Plot Boxplot
plt.figure()
data.boxplot()

# Title and labels
plt.title("Box Plot Showing Skewness")
plt.ylabel("Values")

plt.show()
```

```python
print(data.skew())
```
