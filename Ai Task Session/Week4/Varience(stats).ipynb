```python
import numpy as np
d=[207,230,33,44]
sample_varience=np.var(d,ddof=1)
print("sample_Varience = ",sample_varience)
population_Varence=np.var(d,ddof=0)
print("population_Varience = ",population_Varence)
```

```python
import pandas as pd
d= [207,230,33,44]
series=pd.Series(d)
sample_var=series.var()
print("sample_varience = ",sample_var)
population_Var=series.var(ddof=0)
print("population_Varence = ",population_Var)
```

```python
from scipy import stats
d= [207,230,33,44]
sample_var=stats.tvar(d)
print("sample_varience = ",sample_var)
population_Var=stats.tvar(d,ddof=0)
print("population_Varence = ",population_Var)
```
