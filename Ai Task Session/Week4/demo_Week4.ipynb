```python
# Mean(Manual Method)
population = [15,14,13,12,17,16]
sample = [1,9,2,4]

pop_mean = sum(population)/len(population)
print(pop_mean)

sample_mean = sum(sample)/len(sample)
print(sample_mean)
```

```python
# Mean(Using Numpy)
import numpy as np
a= [2,3,4,6]
b=np.mean(a)
print(b)
```

```python
# Mean(Using Satastic)
import statistics as stat
# Mode(using stat)
var =[4,77,12,8,44,12,44]
var_mean=stat.mean(var)
print("Mean",var_mean)

var_mode=stat.multimode(var)
var_mode2=stat.mode(var)
print("MultiMode",var_mode)
print("Mode",var_mode2)
```

```python
# Mode(using scipy)
from scipy import stats
d =[3,5,77,22,3]
mode =stats.mode(d)
print("Mode using scipy: ",mode)
```
