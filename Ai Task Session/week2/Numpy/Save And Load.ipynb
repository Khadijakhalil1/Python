# *Saving And Loading of Numpy Array*

```python
arr=np.array([1,2,3,4])

np.save('data.npy',arr)
```

```python
arr=np.load('data.npy')
print(arr)
```

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
np.savez("Multiple.npz",a=a,b=b)
```

```python
data=np.load("Multiple.npz")
print(data["a"])
print(data["b"])
```
