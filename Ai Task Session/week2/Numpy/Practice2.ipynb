# *Data Loading and saving*

```python
a=[1,2,3,4]
arr=np.array(a)
print(arr)
```

```python
np.save("numpydat2.npy",arr)
```

```python
arr=np.load("numpydat2.npy")
```

```python
print(arr)
```

# *Multiple array save and load.npz*

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
np.savez("MultipleArray.npz",a=a,b=b)
```

```python
c=np.load("MultipleArray.npz")
print(c["a"])
print(c["b"])
```
