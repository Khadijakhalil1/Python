# *Dimension*


```python
j=np.array([3,6,8])
print(j.ndim)
j
j=np.array([[3,6,8],[1,2,3]])
print(j.ndim)
```

```python
j=np.array([[[3,6,8],[1,2,7]]])
print(j.ndim)  #2x3
j
```

```python
j=np.array([[3,6],[1,2],[9,8]])
print(j.ndim)     #3x2
j
```

```python
#khud sy dimension bnanay k liy
s = [1,2,3,4]   #first list li ha then
arr = np.array(s,ndmin=3)
print(arr.ndim)
arr
```

```python
t = [1,2,3,4]
arr = np.array(t,ndmin=3)
print(arr.ndim)
arr
```
