```python
# num = int(input("Enter number: "))
# reverse = 0
# while num > 0:
#     digit = num % 10
#     reverse = reverse * 10 + digit
#     num = num // 10
# print("Reversed number is:", reverse)
n = int(input("Enter num : "))
rev=0
while n > 0:
    d = n % 10
    rev = rev * 10 + d
    n = n // 10
print("Reverse num is",rev)  
```

```python
# prime or not 
n= int(input("Enter num : "))
if n<=0:
    print("Not prime")
else:
    for i in range(2,n):
        if n %i ==0:
            print("not prime : ")
            break
    else:
        print("Prime")
```
