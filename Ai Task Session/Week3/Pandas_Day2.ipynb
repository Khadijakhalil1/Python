# *Filling Missing Values*

#####  *Fixed Values Se Fill*

```python
df["Marks"].fillna(0,inplace=True)
```

##### fill by mean

```python
df["Marks"].fillna(df["Marks"].mean(),inplace=True)
```

#### Fill by Median

```python
df["Marks"].fillna(df["Marks"].median(),inplace =True)
```

#### Fill By Mode(Categorical Data)

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Mona", "Saira", "Ali", "Hina"],
    "Marks": [85, np.nan, 78, None],
    "City": ["Lahore", "Karachi", None, "FSD"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["City"].fillna(df["City"].mode()[0])
```
