# Plotting

```python
import numpy as np
import pandas as pd
data = {
"Name":["Furqan","Sultan","Faizan","Noman"],
    "Department":["IT","HR","HR","IT"],
    "Salary":[60000,50000,73344,4000]
}
df = pd.DataFrame(data)
df
```

```python
df.plot()
```

### Line Plot

```python
df.plot(kind="line")
```

### Bar plot

```python
df.plot(kind="bar")
```

### Horizontal Bar

```python
df.plot(kind ="barh")
```

### Histogram

```python
df.plot(kind ="hist")
```

### Box plot

```python
df.plot(kind= "box" )
```

### scatter plot

```python
df.plot(kind = "scatter", x="Department" ,y="Salary")
```

### Area plot

```python
df.plot(kind ="area")
```

## Density/KDE

```python
df["Salary"].plot(kind="kde")
```

### pie plot

```python
df["Department"].value_counts().plot(kind ="pie")
```

### Hexbin plot

```python
df.plot(kind ="hexbin",x="Salary",y="Salary",gridsize=20)
```
