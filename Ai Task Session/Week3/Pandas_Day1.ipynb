# Missing value/Data

```python
import pandas as pd
import numpy as np
```

```python
data={
    "name":["Ali","Sultan","Horia","Aniya","Hadi","Noor"],
    "Marks":[85,np.nan,78,None,65,0], 
     "City":["LHR","SHK","karachi","islamabad","FSD",None]
}
df=pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna()  #drop row which contain nan
```

```python

df.dropna(subset=["Marks"])   # drop marks which contain nan value
```

# Filling Missing Values

### Fixed Value

```python
data1={
    "Name":["Hania","Bilal","Shazia","Noor"],
    "Marks":[33,66,None,96],
    "City":["noorpur","SHK","LHR","FSD"]
}
```

```python
df2=pd.DataFrame(data1)
df2
```

```python
df2["Marks"].fillna(10)
```

### Fill by Mean

```python
df2["Marks"].fillna(df["Marks"].mean())
```

```python
df2["City"].fillna(df["City"].mode()[0],inplace=True)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Hina","CR","Monkey"],
    "Marks": [85, np.nan, 78, None,65,0],
    "City": ["Lahore", "Karachi", None, "Islamabad","Lahore","Jungle"]
}

df3 = pd.DataFrame(data)
df3
```

### Fill By mode(Categorical Data)

```python
df3["City"].fillna(df3["City"].mode()[0], inplace=True)
```

### Fill by median


```python
df2
```

```python
df2["Marks"].fillna(df2["Marks"].median())
```

### Forward Fill(Copy Privous value)

```python
df.ffill()
```

### Backward Fill

```python
df.bfill()
```
