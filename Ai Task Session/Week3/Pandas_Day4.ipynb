# Concatenation

```python
import pandas as pd
data1= {
    "ID":[1,2,3,4,5,6,7],
     "Name" :["Khadija","mona","mala","fazi","khadi","hadi","aniya"]
   }
df1 =pd.DataFrame(data1)
df1
```

```python
data2={
    "ID":[11,25,13,4,5,63,77],
     "Marks":[76,88,69,86,92,68,95]
}
df2=pd.DataFrame(data2)
df2
```

### column wise conactenation

```python
pd.concat([df1,df2],axis=1)
```

### Row wise conactenation

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=0)  #by default row wis conact
```

# Groupby

```python
import pandas as pd 
import numpy as np
data={
    "Name":["Faizan","Khadija","Aniya","Sultan"],
    "Department":["IT","HR","HR","IT",],
    "Salary":[90000,70000,50000,80000]
}
df =pd.DataFrame(data)
df
```

```python
df.groupby("Department")["Salary"].mean() #Department wise average salary
```

```python
df.groupby("Department")["Salary"].agg(["mean","max","min","count"]) #multiple Agg
```

```python
df.groupby(["Department","Name"]).sum()
```

# Discretization & Binning

```python
ages =[12,18,25,30,45,70,5]
df =pd.DataFrame({"Age":ages})
```

###  Equal width Binning(pd.cut)

```python
df["Age_Group"]=pd.cut(df["Age"],
                      bins=[0,18,30,50,80],
                      labels=["Teen","Young","Adult","Senoir"])
```

```python
df
```

### Equal frequency Binning(pd.qcut)

```python
df["Age_Group"] =pd.qcut(df["Age"],q=3)
```

```python
df
```
