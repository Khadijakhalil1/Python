```python
import pandas as pd
import numpy as np
```

##  *Merging*

```python
data1= {
    "ID":[1,2,3,4,5,6,7],
     "Name" :["Khadija","mona","mala","fazi","khadi","hadi","aniya"]
   }
df1 =pd.DataFrame(data1)
df1
```

```python
data2={
    "ID":[11,25,13,4,5,63,77],
     "Marks":[76,88,69,86,92,68,95]
}
df2=pd.DataFrame(data2)
df2
```

```python
df = pd.merge(df1,df2,on="ID")
```

```python
df
```

# Types of Merging

## 1.Inner Join

```python
pd.merge(df1,df2,on="ID",how="inner") #only contain common ID
```

## outer join

```python
pd.merge(df1,df2,on="ID",how="outer")   #1.contian all data  2.only unique values 3.left+right-common
```

## Left join

```python
pd.merge(df1,df2,on="ID",how="left")
```

## Right Join

```python
df1.join(df2,lsuffix="_left",rsuffix="_right")
```
